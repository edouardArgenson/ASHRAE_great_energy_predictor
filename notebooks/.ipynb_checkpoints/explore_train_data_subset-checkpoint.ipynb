```python
# Load building_metadata and set index
bdata = pd.read_csv('../data/raw/csvs/building_metadata.csv', 
                    usecols=['site_id', 'building_id'], index_col='building_id')
bdata.head()
```

```python
# Keep only buildings we haven't kicked out of df
alive_buildings = df.index.get_level_values('building_id').unique()
all_buildings = bdata.index.unique()
buildings_to_drop = all_buildings.difference(alive_buildings)
bdata.drop(buildings_to_drop, inplace = True)
bdata.head()
```

```python
print('Number of site 0 buildings in our subset : {}'.format(bdata['site_id'].value_counts()[0]))
```

```python
# Join by building_id
dfs = df.join(bdata, on='building_id', how='left')
print('non-found site_ids : {}'.format(dfs['site_id'].isna().sum()))
dfs.head()
```

```python
# kick out site_ids == 0
dfs_w0 = dfs[dfs['site_id'] != 0]
dfs_w0.head()
```

```python
# TODO change colors
```

```python
# Randomly select a building
random.seed(1)
building_list = dfs_w0.index.get_level_values('building_id').unique().tolist()
sampled_id = random.choice(building_list, 1)[0]
print(sampled_id)
```

```python
to_plot = dfs_w0.loc[sampled_id]
to_plot.head()
```

```python
# All year

sns.set_style("whitegrid")
sns.set_palette(sns.color_palette("muted"))

fig = plt.figure(figsize=(16,4))
with sns.axes_style("whitegrid"):
    ax = sns.lineplot(x="timestamp", y="meter_reading", hue="meter",
    data=to_plot.reset_index(inplace=False))
    ax.xaxis.set_major_locator(mdates.MonthLocator())
    #ax.xaxis.set_minor_locator(mdates.HourLocator())
    plt.xticks(rotation=50, horizontalalignment='right')
```

```python
# Fewer months

start = np.datetime64('2016-03-01')
stop = np.datetime64('2016-05-10')
to_plot_interval = df_plt = to_plot.loc[start:stop]

fig = plt.figure(figsize=(16,4))
with sns.axes_style("whitegrid"):
    ax = sns.lineplot(x="timestamp", y="meter_reading", hue="meter",
    data=to_plot_interval.reset_index(inplace=False))
    ax.xaxis.set_major_locator(mdates.MonthLocator())
    #ax.xaxis.set_minor_locator(mdates.HourLocator())
    plt.xticks(rotation=50, horizontalalignment='right')
```

```python
# Fewer months

start = np.datetime64('2016-03-01')
stop = np.datetime64('2016-03-20')
to_plot_interval = df_plt = to_plot.loc[start:stop]

fig = plt.figure(figsize=(16,4))
with sns.axes_style("whitegrid"):
    ax = sns.lineplot(x="timestamp", y="meter_reading", hue="meter",
    data=to_plot_interval.reset_index(inplace=False))
    ax.xaxis.set_major_locator(mdates.DayLocator())
    plt.xticks(rotation=50, horizontalalignment='right')
```
